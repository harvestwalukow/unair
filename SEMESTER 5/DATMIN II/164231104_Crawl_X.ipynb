# 2. Wordcloud & Clustering

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import warnings
warnings.filterwarnings('ignore')
```

### Wordcloud

```python
# Load data
file_path = "tweets-data/PRABOWO.csv"
df = pd.read_csv(file_path, delimiter=",")

# Text cleaning function
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Stopwords Indonesia
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
indonesian_stopwords = set(factory.get_stop_words())

additional_stopwords = {
    'pidato', 'prabowo'
}
indonesian_stopwords.update(additional_stopwords)

def simple_tokenize(text):
    return text.split()

def process_text(text):
    cleaned_text = clean_text(text)
    tokens = simple_tokenize(cleaned_text)
    filtered_tokens = [word for word in tokens if word not in indonesian_stopwords and len(word) > 2]
    return ' '.join(filtered_tokens)

df['cleaned_text'] = df['full_text'].apply(process_text)

all_text = ' '.join(df['cleaned_text'].dropna())

plt.figure(figsize=(15, 10))

# Buat wordcloud
wordcloud = WordCloud(
    width=1200,
    height=800,
    background_color='white',
    max_words=100,
    colormap='viridis',
    relative_scaling=0.5,
    random_state=42
).generate(all_text)

# Display wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Wordcloud dari Tweet tentang Prabowo', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

### Clustering

```python
# Import Libraries untuk Clustering
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Membuat matrix TF-IDF 
tfidf_vectorizer = TfidfVectorizer(
    max_features=500,
    min_df=2,
    max_df=0.7,
    stop_words=None,
    ngram_range=(1, 2)
)

tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_text'])
print(f"Shape TF-IDF matrix: {tfidf_matrix.shape}")
```

```python
# Cari jumlah cluster optimal
silhouette_scores = []
K_range = range(2, 7)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(tfidf_matrix)
    silhouette_scores.append(silhouette_score(tfidf_matrix, cluster_labels))

# Plot Silhouette Score
plt.figure(figsize=(8, 5))
plt.plot(K_range, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score untuk Optimal k')
plt.grid(True)
plt.show()

# Pilih k optimal
optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"Jumlah cluster optimal: {optimal_k}")
```

```python
# Clustering final
print(f"Melakukan clustering dengan {optimal_k} cluster...")
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(tfidf_matrix)

df['cluster'] = cluster_labels

# Analisis hasil clustering
print(f"\nDistribusi cluster:")
cluster_counts = df['cluster'].value_counts().sort_index()
for cluster_id, count in cluster_counts.items():
    print(f"Cluster {cluster_id}: {count} tweets ({count/len(df)*100:.1f}%)")
```

```python
# Analisis kata kunci per cluster
print(f"Kata kunci per cluster:")
feature_names = tfidf_vectorizer.get_feature_names_out()

for cluster_id in range(optimal_k):
    print(f"\n--- Cluster {cluster_id} ---")
    
    # Ambil tweets dari cluster ini
    cluster_mask = cluster_labels == cluster_id
    cluster_tfidf = tfidf_matrix[cluster_mask]
    
    # Hitung rata-rata TF-IDF untuk cluster ini
    cluster_mean = np.array(cluster_tfidf.mean(axis=0)).flatten()
    
    # Ambil top 5 kata dengan TF-IDF tertinggi
    top_indices = cluster_mean.argsort()[-5:][::-1]
    top_words = [feature_names[i] for i in top_indices]
    
    print("Top 5 kata kunci:")
    for word in top_words:
        print(f"  {word}")
    
    # Tampilkan contoh tweet
    cluster_tweets = df[df['cluster'] == cluster_id]['full_text'].head(2)
    print(f"\nContoh tweet:")
    for i, tweet in enumerate(cluster_tweets, 1):
        print(f"  {i}. {tweet[:80]}...")
```

```python
# Visualisasi clustering dengan PCA
print(f"Membuat visualisasi 2D dengan PCA...")
pca = PCA(n_components=2, random_state=42)
tfidf_2d = pca.fit_transform(tfidf_matrix.toarray())

plt.figure(figsize=(10, 6))
colors = plt.cm.Set3(np.linspace(0, 1, optimal_k))

for cluster_id in range(optimal_k):
    mask = cluster_labels == cluster_id
    plt.scatter(tfidf_2d[mask, 0], tfidf_2d[mask, 1], 
               c=[colors[cluster_id]], label=f'Cluster {cluster_id}', 
               alpha=0.7, s=50)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.title('Visualisasi Clustering Tweet dengan PCA')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nSilhouette Score: {silhouette_score(tfidf_matrix, cluster_labels):.3f}")
print(f"Variance explained by PC1 + PC2: {sum(pca.explained_variance_ratio_):.2%}")
```

# 3. Social Network

```python
import networkx as nx
```

```python
# Ekstrak interaksi pengguna dari tweets
def extract_interactions(df):
    interactions = []
    
    for idx, row in df.iterrows():
        tweet_text = str(row['full_text'])
        user_id = row['user_id_str']
        
        # Ekstrak mentions (@username)
        mentions = re.findall(r'@(\w+)', tweet_text)
        
        # Tambahkan interaksi balasan jika ada
        if pd.notna(row['in_reply_to_screen_name']) and row['in_reply_to_screen_name'] != 'None':
            interactions.append({
                'from_user': user_id,
                'to_user': row['in_reply_to_screen_name'],
                'type': 'reply'
            })
        
        # Tambahkan interaksi mention
        for mention in mentions:
            interactions.append({
                'from_user': user_id,
                'to_user': mention,
                'type': 'mention'
            })
    
    return pd.DataFrame(interactions)

# Buat jaringan interaksi
interactions_df = extract_interactions(df)

print(f"Total interactions: {len(interactions_df)}")
print(f"Reply interactions: {len(interactions_df[interactions_df['type'] == 'reply'])}")
print(f"Mention interactions: {len(interactions_df[interactions_df['type'] == 'mention'])}")

# Buat graf NetworkX
G = nx.DiGraph()

# Tambahkan edges (interaksi)
for _, interaction in interactions_df.iterrows():
    G.add_edge(interaction['from_user'], interaction['to_user'])

print(f"Network created: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")
```

# 4. Centrality Measures

### Degree Centrality

```python
# Hitung jumlah mention yang diterima setiap user
in_degree = dict(G.in_degree())

# Hitung jumlah interaksi yang dilakukan setiap user  
out_degree = dict(G.out_degree())

# Buat dataframe untuk top 5 user yang paling banyak di-mention
top_mentioned_df = pd.DataFrame(in_degree.items(), columns=['User', 'Mentions_Received'])
top_mentioned_df = top_mentioned_df.sort_values('Mentions_Received', ascending=False).head(5)

# Buat dataframe untuk top 5 user yang paling aktif berinteraksi
top_active_df = pd.DataFrame(out_degree.items(), columns=['User', 'Interactions_Made']) 
top_active_df = top_active_df.sort_values('Interactions_Made', ascending=False).head(5)
```

### Closeness Centrality

```python
# Hitung closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Buat dataframe untuk top 5 user dengan closeness centrality tertinggi
top_closeness_df = pd.DataFrame(closeness_centrality.items(), columns=['User', 'Closeness_Centrality'])
top_closeness_df = top_closeness_df.sort_values('Closeness_Centrality', ascending=False).head(5)

top_closeness_df
```

### Betweenness Centrality

```python
# Hitung betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Buat dataframe untuk top 5 user dengan betweenness centrality tertinggi
top_betweenness_df = pd.DataFrame(betweenness_centrality.items(), columns=['User', 'Betweenness_Centrality'])
top_betweenness_df = top_betweenness_df.sort_values('Betweenness_Centrality', ascending=False).head(5)

top_betweenness_df
```

### Eigenvector Centrality

```python
# Hitung eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# Buat dataframe untuk top 5 user dengan eigenvector centrality tertinggi
top_eigenvector_df = pd.DataFrame(eigenvector_centrality.items(), columns=['User', 'Eigenvector_Centrality'])
top_eigenvector_df = top_eigenvector_df.sort_values('Eigenvector_Centrality', ascending=False).head(5)

top_eigenvector_df
```

# 5. Visualisasi Network

```python
# Create clear visualization
plt.figure(figsize=(15, 12))

# Better layout
pos = nx.spring_layout(G, k=2, iterations=50, seed=42)

# Node sizes - make them visible
if G.is_directed():
    node_sizes = [G.in_degree(node) * 800 + 300 for node in G.nodes()]
    centrality_type = "In-Degree (Mentions Received)"
else:
    node_sizes = [G.degree(node) * 600 + 200 for node in G.nodes()]
    centrality_type = "Degree (Connections)"

# Draw edges
if G.is_directed():
    nx.draw_networkx_edges(G, pos, alpha=0.3, edge_color='gray',
                          arrows=True, arrowsize=12, arrowstyle='->', width=1)
else:
    nx.draw_networkx_edges(G, pos, alpha=0.3, edge_color='gray', width=1)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=node_sizes,
                      node_color='lightblue', alpha=0.8, edgecolors='black', linewidths=1.5)

# Add labels for top 5 users only
if G.is_directed():
    top_5 = sorted(G.in_degree(), key=lambda x: x[1], reverse=True)[:5]
else:
    top_5 = sorted(G.degree(), key=lambda x: x[1], reverse=True)[:5]

top_5_labels = {user[0]: f"User {str(user[0])[:8]}..." for user in top_5}
nx.draw_networkx_labels(G, pos, top_5_labels, font_size=10, font_weight='bold')

# Title
network_type = "Interaction Network" if G.is_directed() else "Similarity Network"
plt.title(f'{network_type}\\n(Node size = {centrality_type})', 
         fontsize=16, fontweight='bold', pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```
